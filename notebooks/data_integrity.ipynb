```python
import os
from dotenv import load_dotenv
import pandas as pd
from sqlalchemy import create_engine

# 1) Carga las vars del .env
load_dotenv()  

user     = os.getenv('DB_USER')
password = os.getenv('DB_PASSWORD')
host     = os.getenv('DB_HOST')
port     = os.getenv('DB_PORT')
db       = os.getenv('DB_NAME')
table_name = os.getenv('TABLE_NAME')

# 2) Crea la URL de conexión
engine = create_engine(
    f'mysql+pymysql://{user}:{password}@{host}:{port}/{db}'
)

# 3) Lee todas las tablas de dataset_db
try:
    data = pd.read_sql(f'SELECT * FROM {table_name};', con=engine)
    
except Exception as e:
    print(f'Error al leer las tablas: {e}')
    data = pd.DataFrame()
    
finally:
    display(data)
```

```python
from deepchecks.tabular import Dataset
from deepchecks.tabular.suites import data_integrity

# 1) Crear el dataset
dataset = Dataset(
    df=data,
    label='target', 
    cat_features=['cat1', 'cat2', 'disc1', 'disc2']
)

# 2) Crea y ejecuta la suite de integridad
result = data_integrity().run(dataset)   # corre todos los checks

# 3) Muestra el informe
result.show()
```
